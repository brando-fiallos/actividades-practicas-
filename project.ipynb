```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
df.head()

```

```python
setosa = df[df['species'] == 'setosa']
df['petal_ratio'] = df['petal_length'] / df['petal_width']
df.groupby('species').mean()
```

```python

sepal_lengths = df['sepal_length'].values
print("Media:", np.mean(sepal_lengths))
print("Mediana:", np.median(sepal_lengths))
print("Desviaci칩n est치ndar:", np.std(sepal_lengths))
```

```python
plt.hist(df['petal_length'], bins=20, color='lightgreen')
plt.title("Histograma de Longitud del P칠talo")
plt.xlabel("Longitud")
plt.ylabel("Frecuencia")
plt.show()


df['species'].value_counts().plot(kind='bar', color='coral')
plt.title("Cantidad de registros por especie")
plt.xlabel("Especie")
plt.ylabel("Cantidad")
plt.show()


colors = {'setosa':'red', 'versicolor':'blue', 'virginica':'green'}
for specie in df['species'].unique():
    temp = df[df['species'] == specie]
    plt.scatter(temp['sepal_length'], temp['petal_length'], label=specie, color=colors[specie])
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Relaci칩n entre Sepal y Petal Length")
plt.legend()
plt.show()


```

```python

fig, axs = plt.subplots(1, 2, figsize=(12, 5))
axs[0].hist(df['sepal_length'], color='skyblue')
axs[1].hist(df['petal_length'], color='orange')
axs[0].set_title("Longitud del s칠palo")
axs[1].set_title("Longitud del p칠talo")
plt.tight_layout()
plt.show()

```

```python
media = np.mean(df['petal_width'])
mediana = np.median(df['petal_width'])
desviacion = np.std(df['petal_width'])

print("\n--- Estad칤sticas de petalo ---")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviaci칩n est치ndar: {desviacion:.2f}")
```

```python
plt.figure(figsize=(8, 5))
promedios = df.groupby('species')['sepal_width'].mean()
promedios.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Promedio de sepalo por especie', fontweight='bold')
plt.xlabel('Especie')
plt.ylabel('Promedio sepalo (cm)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
filtrados = df[df['sepal_length'] > 6.0]
plt.hist(filtrados['petal_length'], bins=12, color='purple', edgecolor='white')
plt.title('Distribuci칩n de petalo (sepalo > 6.0 cm)', fontweight='bold')
plt.xlabel('Longitud del p칠talo (cm)')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
colores = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}

for especie, grupo in df.groupby('species'):
    plt.scatter(
        grupo['sepal_width'], 
        grupo['petal_width'], 
        color=colores[especie], 
        label=especie,
        alpha=0.7
    )

plt.title('Relaci칩n entre sepalo y petalo', fontweight='bold')
plt.xlabel('Ancho del s칠palo (cm)')
plt.ylabel('Ancho del p칠talo (cm)')
plt.legend()
plt.grid(linestyle='--', alpha=0.3)
plt.show()
```

```python
df['petal_ratio'] = df['petal_length'] / df['petal_width']
df.to_csv('iris_modificado.csv', index=False)
print("\n--- Archivo guardado ---")
print("DataFrame guardado como 'iris_modificado.csv'")
```
